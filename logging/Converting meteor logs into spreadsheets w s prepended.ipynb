```python
spreadsheetRows = {}
with open('final_q4_log.txt','r') as g:
    for line in g:
        entry_array = [tok.rstrip('\n') for tok in line.split(',')]
        print entry_array
        if entry_array[0].startswith('s'):
            entry_array[0] = entry_array[0][1:]
        print entry_array
        spreadsheetRows[int(entry_array[0])]=tuple(entry_array[1:])
#print spreadsheetRows

import datetime 
with open('final_q4_processed.csv','a') as h:
    for key in range(int(sorted(spreadsheetRows)[-1])+1):
        #print 'row',row
        key_str = str(key)
        if key in spreadsheetRows.keys():
            h.write(key_str+','+spreadsheetRows[key][0]+','+spreadsheetRows[key][1])
        else:
            h.write(key_str)
        h.write('\n')
```
